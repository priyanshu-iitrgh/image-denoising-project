```python
import cv2
import numpy as np
import glob
import os

def calculate_psnr(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if mse == 0: 
        return float('inf')
    max_pixel = 255.0
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr


original_dir = r'C:\Users\priyanashu\OneDrive\Desktop\VLG\Train\high'
low_light_dir = r'C:\Users\priyanashu\OneDrive\Desktop\VLG\Train\denoised'


original_images = sorted(glob.glob(os.path.join(original_dir, '*')))
low_light_images = sorted(glob.glob(os.path.join(low_light_dir, '*')))


assert len(original_images) == len(low_light_images), "The number of images in both directories should be the same."


psnr_values = []

for original_image_path, low_light_image_path in zip(original_images, low_light_images):
    
    original_image = cv2.imread(original_image_path)

    
    low_light_image = cv2.imread(low_light_image_path)

    
    assert original_image.shape == low_light_image.shape, f"Image dimensions do not match for {original_image_path} and {low_light_image_path}"

    
    psnr_value = calculate_psnr(original_image, low_light_image)
    psnr_values.append(psnr_value)
    print(f"PSNR value for {os.path.basename(original_image_path)} is {psnr_value} dB")


average_psnr = np.mean(psnr_values)
print(f"Average PSNR value for all images is {average_psnr} dB")
```
